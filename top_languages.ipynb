```python
import pandas as pd
import matplotlib

languages_xls = pd.read_excel("Top+Ten+Refugee+Native+Languages(10.2.18).xls")
languages_xls
```

```python
languages_xls = languages_xls[10:21]
languages_xls
```

```python
languages_xls.columns = languages_xls.iloc[0]
languages_xls
```

```python
languages_xls = languages_xls.reindex(languages_xls.index.drop(10))
languages_xls
```

```python
languages_xls = languages_xls.dropna(axis="columns")
```

```python
languages_graph = languages_xls[languages_xls.columns[1:13]]
languages_graph
```

```python
languages_graph = languages_graph.transpose()
languages_graph
```

```python
languages_graph.columns = languages_graph.iloc[0]
languages_graph
```

```python
languages_graph = languages_graph.drop(languages_graph.index[0])
languages_graph
```

```python
import pandas as pd
import plotly as plt
import plotly.plotly as py
import plotly
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import helpers
import numpy as np 
plotly.offline.init_notebook_mode(connected=True)

languages = languages_graph.columns
traces = []
for language in languages:
    trace = go.Scatter(
        x = languages_graph.index,
        y = languages_graph[language],
        mode = 'lines',
        name = language
    )
    traces.append(trace)
    
    

layout = go.Layout(
    title='Refugee admission by native language',
    xaxis=dict(title='Year'),
    yaxis=dict(title='Count'),
    margin=go.Margin(
        l=50,
        r=50,
        b=200,
        t=50,
        pad=4
    ),
)


fig = go.Figure(data=traces,layout=layout)
iplot(fig)
```
