```python
import pandas as pd
```

```python
welcome_cities_df = pd.read_csv("Welcoming America Cities.csv")
```

```python
welcome_cities_df.head()
```

```python
welcome_cities_df = welcome_cities_df[["city", "state"]]
welcome_cities_df.head()
```

```python
xls = pd.ExcelFile("2009-2013-acs-lang-tables-cbsa.xls")
xls_cities = list(xls.sheet_names)
```

```python
xls_cities
```

```python
selected_cities = []
```

```python
welcome_cities_list = welcome_cities_df["city"].tolist()
welcome_cities_list
```

```python
selected_cities = []
for welcome_city in welcome_cities_list:
    # Adds to selected_cities if any city matches xls_cities
    for xls_city in xls_cities:
        if welcome_city in xls_city:
            selected_cities.append(xls_city)
```

```python
selected_cities = list(set(selected_cities))
selected_cities
```

```python
len(selected_cities)
```

```python
selected_cities
```

```python
sheets = []
for sheet in selected_cities:
    df = xls.parse(sheet)
    df["City"] = sheet
    sheets.append(df)
mass_df = pd.concat(sheets)
mass_df
```

```python
mass_df.columns = ["Languages", "# of Speakers", "1", "2", "3", "City"]
```

```python
mass_df = mass_df[["Languages", "# of Speakers", "City"]]
mass_df.head()
```

```python
def remove_dots(text):
    if text != None and type(text) != float:
        return text.replace('.', '')
    else:
        return text
    
mass_df['Languages'] = mass_df['Languages'].apply(remove_dots)
```

```python
mass_df.head(20)
```

```python
def select_rows_with_languages_that_we_want(language_string):
    if language_string in ["Spanish", "Burmese", "Swahilli", "Karen", "Persian", "Armenian", "Arabic", "Nepali", "French", "Urdu"]:
        return True
    else:
        return False
    
mass_df = mass_df[mass_df.Languages.apply(select_rows_with_languages_that_we_want)]
mass_df
```

```python
mass_df = mass_df.set_index("Languages")
mass_df
```

```python
mass_df = mass_df.groupby(["City", "Languages"]).sum()
mass_df
```

```python
mass_df = mass_df.reset_index()
mass_df
```

```python
mass_df.isnull().values.any()
```

```python
import numpy as np

mass_df = mass_df.replace("(D)", np.NaN)
```

```python
mass_df
```

```python
mass_df["# of Speakers"] = mass_df["# of Speakers"].apply(pd.to_numeric)
```

```python
mass_df.to_csv("welcome_city_languages.csv")
```
